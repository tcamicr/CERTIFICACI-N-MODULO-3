## 5. Avoid Using Global Variables

```python
# What Not to Do
total = 0
def add_to_total(numbers):
    global total
    for number in numbers:
        total += number
```

## 15. Use `multiprocessing` for Parallel Processing

## 17. Use `lru_cache` for Caching

```python
# What Not to Do
def fibonacci(n):
    if n < 2:
        return n
    return fibonacci(n-1) + fibonacci(n-2)
```

```python
from functools import lru_cache

# Optimized Version (using lru_cache)
@lru_cache(maxsize=None)
def fibonacci(n):
    if n < 2:
        return n
    return fibonacci(n-1) + fibonacci(n-2)
```

## 20. Optimize Algorithms

```python
# What Not to Do (linear search)
def linear_search(arr, x):
    for i in range(len(arr)):
        if arr[i] == x:
            return i
    return -1
```

```python
# Optimized Version (binary search)
def binary_search(arr, x):
    low, high = 0, len(arr) - 1
    while low <= high:
        mid = (low + high) // 2
        if arr[mid] < x:
            low = mid + 1
        elif arr[mid] > x:
            high = mid - 1
        else:
            return mid
    return -1
```

## 21. Use `@staticmethod` and `@classmethod`

## 22. Use `slots` to Reduce Memory Overhead

```python
# What Not to Do
class MyClass:
    def __init__(self, attr1, attr2):
        self.attr1 = attr1
        self.attr2 = attr2
```

```python
# Optimized Version (using slots)
class MyClass:
    __slots__ = ['attr1', 'attr2']
    def __init__(self, attr1, attr2):
        self.attr1 = attr1
        self.attr2 = attr2
```

## 23. Avoid Using `try`/`except` in Performance-Critical Code

```python
# What Not to Do
def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return None
```

```python
# Optimized Version
def safe_divide(a, b):
    if b == 0:
        return None
    return a / b
```

## 25. Use `contextlib` for Context Managers

```python
# What Not to Do
class MyContext:
    def __enter__(self):
        print("Entering")
    def __exit__(self, exc_type, exc_value, traceback):
        print("Exiting")

with MyContext():
    print("Inside")
```

```python
from contextlib import contextmanager

# Optimized Version
@contextmanager
def my_context():
    print("Entering")
    yield
    print("Exiting")

with my_context():
    print("Inside")
```

## 26. Use `asyncio` for Asynchronous I/O

## 30. Use `heapq` for Priority Queues

```python
# What Not to Do
queue = []
queue.append((2, 'task2'))
queue.append((1, 'task1'))
queue.sort()
smallest = queue.pop(0)
```
